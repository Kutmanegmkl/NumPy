1) Shape and dimension

```python
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])

a.shape #return tuple
a.ndim
a.size

a.size == a.shape[0]*a.shape[1]
```

2) Data types

```python
#dtype
a.dtype
a.dtype.name
a.dtype.itemsize
a.dtype.kind
```

3) Memory 

```python
#itemsize
a.itemsize
np.array([1,2,3], dtype='int8').itemsize
np.array([1,2,3], dtype=np.float32).itemsize

#nbytes
a.nbytes
a.nbytes ==a.size*a.itemsize #always true 

#strides
a.strides #return tuple
b = a.T
b.strides
b.shape
```

4) Memory management

```python
#data(object memoryview)
a.data

#base
a = np.array([1, 2, 3, 4, 5])
b=a[1:4] # b is view

b.base is a
a.base is None

#flags 
a.flags['WRITEABLE']
a.flags['C_CONTIGUOUS']
a.flags['OWNDATA']

# you can forbid changing the array
a.flags['WRITEABLE'] = False
a[0]=12 #ValueError: assignment destination is read-only
```

5) Alternative data representations

```python
# Transposition
a = np.array([[1, 2, 3], [4, 5, 6]])
a.T
a.shape

# for 1D array
v = np.array([1, 2, 3, 4, 5])
v.T.shape

v.reshape(-1, 1).shape

#real & imag 
c = np.array([1+2j, 3+4j, 5+9j])
c.real, c.imag

#real numbers
r = np.array([1, 2, 3, 4])
r.real
r.imag #zeros

#flat (iterator)
a = np.array([[1, 2, 3], [4, 5, 6]])
for i in a.flat:
    print(i)
a.flat[3] #indexing
a.flat[2:5]
a.flat[0]=99 #assignment
a
```
