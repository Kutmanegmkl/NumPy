```python
import numpy as np
```

1) Из существуюущих данных в Python

```python
# from the list

list_1=[1,2,3,4]
list_2=[[1,2,3],[4,5,6]]
a=np.array(list_1)
b=np.array(list_2)

a,b
```

```python
#from the tuple

tuple_=(1,2,3,4)
arr=np.array(tuple_)
arr
```

```python
# Parameters ndarray

#dtype
a=np.array([1,2,3], dtype=np.float64)
a, a.dtype

b=np.array([1.4, 2.33, 4.44], dtype=np.int32)
b, b.dtype


# copy (by default - True)
a=np.array([1,2,3])
a
c=np.array(a, copy=False) # creating view
c is a


#ndmin
d=np.array([1,2,3], ndmin=2)
d.shape # (1,3), not (3,)
```

```python
# np.asarray() - by default, it does not copy data.

a=np.array([1,2,3])
b=np.asarray(a)
b is a
```

2) Массивы заполненные константами

```python
# zeros
np.zeros(5)
np.zeros((2,3))
np.zeros((2,3,4))

#ones
np.ones(4)
np.ones((2,3))
np.ones((2,3,3))


#full
np.full(5,4)
np.full((2,3), 3.14)
```

```python
#empty()

np.empty((3,4))

result=np.empty(100)
for i in range(100):
    result[i]=i*2

result
```

```python
# _like 

a=np.array([[1,2,3],[4,5,6]]) #shape=(2,3), dtype=int64

np.zeros_like(a)
np.ones_like(a)
np.full_like(a, 9)
np.empty_like(a)

```

3) Диапазоны и последовательности

```python
# np.arrange()

np.arange(5)
np.arange(0,11,2)
np.arange(10,0,-1)
np.arange(0,2,0.2)
```

```python
# np.linspace()

np.linspace(0,1,5)
np.linspace(1,10,11)
#endpoint=True
np.linspace(0,1,5, endpoint=False)

#retstep - return step
arr, step=np.linspace(0,2,10, retstep=True)
```

```python
#np.logspace()

np.logspace(0, 4, 4)
np.logspace(0, 1, 5)
```

```python
# np.geomspace()

np.geomspace(2, 32, 5)
```
