```python
import numpy as np
```

```python
# full function signature
np.array(object, dtype = None, copy = True, order = 'K', subok = False, ndmin = 0, like = None)
```

Object

```python
np.array([1, 2, 3])  #list
np.array((1, 2, 3)) #tuple
np.array([[1, 2], [3, 4]])  # nested list
np.array(20) #scalar
np.array(range(1,4)) # any iterable object

#another numpy array
a = np.array([1, 2, 3])
b = np.array(a) #created copy 
b
```

dtype

```python
np.array([1, 2, 3]).dtype # 'int64' - by default
np.array([1, 2, 3], dtype = np.bool_)
np.array([1.4, 3.8, 4.9], dtype = np.int8)
np.array(['1', '2', '3'], dtype = np.int64)
```

copy

```python
a = np.array([1, 2, 3])
b = np.array(a, copy = True) # by default - copy = True
b[0] = 99
a # not changed 

c = np.array(a, copy = False) # create view
c[0] = 99
a # changed!

d = np.array([1, 2, 3], dtype = np.int32)
#e = np.array(d, copy = False, dtype = np.float64) # another data type
e = np.asarray(d, dtype = np.float32)
e is d
```
